```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv(
    "market_data_raw.csv",
    header=[0,1]
)
```

```python
data.head()
```

```python
data.columns = [
    f"{col[0]}_{col[1]}" if "Unnamed" not in col[0]
    else "Date"
    for col in data.columns
]
```

```python
print(data.columns)
data = data.iloc[1:]
```

```python
data.rename(
    columns={"Price_Ticker": "Date"},
    inplace=True
)
```

```python
data.head()
```

```python
data["Date"] = pd.to_datetime(data["Date"])
```

```python
data = data.sort_values("Date")
```

```python
print(data.isnull().sum())
```

```python
data = data.ffill()
```

```python
data.set_index("Date", inplace=True)
```

```python
tickers = ["AAPL", "MSFT", "SPY"]

for ticker in tickers:
    data[f"Log_Return_{ticker}"] = np.log(
        data[f"Close_{ticker}"] /
        data[f"Close_{ticker}"].shift(1)
    )
```

```python
data.dropna(inplace=True)
```

```python
data.head()
```

```python
data.to_csv("market_data_processed.csv")
```
